```python
def plot_data(file_path, color):
    n_estimators = []
    roc_auc = []
    num_leaves = []

    with open(file_path, 'r') as f:
        lines = f.readlines()

    # Estrai i valori di n_estimators, roc_auc e num_leaves
    for line in lines:
        if 'n_estimators' in line:
            n_estimators.append(int(line.split('=')[1]))
        elif 'roc_auc' in line:
            roc_auc.append(float(line.split('=')[1]))
        elif 'num. leaves' in line:
            num_leaves.append(int(line.split('=')[1]))

    markers = {30: 'o', 60: 's', 100: '^'}
    for nl, marker in markers.items():
        plt.plot(np.array(n_estimators)[np.array(num_leaves)==nl], np.array(roc_auc)[np.array(num_leaves)==nl], marker=marker, linestyle='-', color=color)

# Crea la figura
plt.figure()

# Plotta i dati dei tre file
plot_data('/Users/lucatabarroni/Desktop/HEPD/BDT/BDT 2/BDT_dataset_1/parameters.txt', 'b')
plot_data('/Users/lucatabarroni/Desktop/HEPD/BDT/BDT 2/BDT_dataset_2/parameters.txt', 'r')
plot_data('/Users/lucatabarroni/Desktop/HEPD/BDT/BDT 2/BDT_dataset_3/parameters.txt', 'g')
# Crea la legenda per i marker
markers = {30: 'o', 60: 's', 100: '^'}
for nl, marker in markers.items():
    plt.plot([], [], marker=marker, linestyle='', color='k', label=f'num_leaves={nl}')

# Crea la legenda per i colori
colors = {'Dataset 1': 'b', 'Dataset 2': 'r', 'Dataset 3': 'g'}
for file, color in colors.items():
    plt.plot([], [], color=color, label=file)

plt.xlabel('Numero di estimatori')
plt.ylabel('ROC AUC')
plt.title('ROC AUC in funzione del numero di estimatori')
plt.legend(title="Markers and Dataset", loc="upper left", bbox_to_anchor=(1,1))
plt.show()
```
